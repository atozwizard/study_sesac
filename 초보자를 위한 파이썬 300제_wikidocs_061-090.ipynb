```python
temp = ('apple', 'banana', 'cake')
(a, b, c) = temp
print(a, b, c)
a, b, c = temp
print(a, b, c)
"('apple', 'banana', 'cake'),('apple', 'banana', 'cake'),('apple', 'banana', 'cake')"
```

**a, b, c = temp** 이라고 쓰는 순간, 바구니 안에 있는 내용물을 순서대로 꺼내서 
왼쪽의 변수들에 딱 하나씩 할당한다.
a 에는 첫 번째인 apple
b 에는 두 번째인 banana
c 에는 세 번째인 cake 
따라서 print(a,b,c)를 하면 각각의내용물인 apple banana cake이 출력된다

(2,4,6,8, ..., 98)
1부터 99까지의 정수 중 짝수만 저장된 튜플 생성

```python
tupl = tuple(range(2,100,2))
print(tupl)
```
