```python
import requests
btc = requests.get("https://api.bithumb.com/public/ticker/").json()['data']
print(btc)
print(type(btc['max_price']))
```

```python
if int(btc['max_price'])  <  int(btc['max_price']) - int(btc['min_price']) + int(btc['opening_price']):
    print('상승장')
else : print('하락장')
```

```python
최고가 = int(btc['max_price'])
변동폭 = int(btc['max_price']) - int(btc['min_price'])
시가 = int(btc['opening_price'])
if (시가 + 변동폭) > 최고가:
    print('상승장')
else : print('하락장')
```
