```python
license_plate = "24가 2210"
print(license_plate[-4],license_plate[-3],license_plate[-2],license_plate[-1])
print(license_plate[-4:])
#슬라이싱에서 시작 인덱스를 생략하면 0으로 간주하고, 끝 인덱스를 생략하면 문자열의 끝을 의미함
```

```python
string = "홀짝홀짝홀짝"
print(string[0::2])
print(string[::2])
#슬라이싱에서 시작 인덱스를 생략하면 0으로 간주하고, 끝 인덱스를 생략하면 문자열의 끝을 의미함
```

```python
string = "PYTHON"
print(string[::-1]) 
#시작, 끝 인덱스를 생략해서 끝과 끝으로 범위를 주고, -1로 역순시퀀스
```
