```python
my_list = ["가", "나", "다", "라"]
print(my_list[3],my_list[2])
print(my_list[2],my_list[1])
print(my_list[1],my_list[0])
for i in [3,2,1]:
    print(my_list[i],my_list[i-1])
for i in [2,1,0]:
    print(my_list[i+1],my_list[i])
for i in [0,1,2]:
    print(my_list[3-i],my_list[2-i])
    
for i in range(len(my_list)-1):
    print(my_list[3-i],my_list[2-i])
```
