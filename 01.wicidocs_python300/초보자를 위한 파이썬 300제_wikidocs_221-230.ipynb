```python
def print_even(list):
    for i in list:
        if i % 2 == 0 :
            print(i)

print_even ([1, 3, 2, 10, 12, 11, 15])
```

```python
def print_even(my_list):
    for i in range(len(my_list)):
        if my_list[i] % 2 == 0:
            print(my_list[i])
print_even ([1, 3, 2, 10, 12, 11, 15])
```

```python
def print_keys(my_dict):
    for keys in my_dict.keys():
        print(keys)
print_keys ({"이름":"김말똥", "나이":30, "성별":0})
```

```python
def print_keys(my_dict):
    for values in my_dict.values():
        print(values)
print_keys ({"이름":"김말똥", "나이":30, "성별":0})
```

```python

def print_5xn(string):    
# s = "아이엠어보이유알어걸" 
    chunk = int(len(string)/5)
    for x in range(chunk + 1):
        print(string[x*5:x*5+5])

print_5xn("아이엠어보이유알어걸오마이프렌마이러브")
```

```python

def printmxn(string,num):    
# s = "아이엠어보이유알어걸" 
    chunk = int(len(string)/num)
    for x in range(chunk+1):
        print(string[x*num:x*num+num])


printmxn("아이엠어보이유알어걸", 3)
```

```python
def my_print (a, b) :
    print("왼쪽:", a)
    print("오른쪽:", b)

my_print(a=100, b=200)

# 왼쪽: 100
# 오른쪽: 200
```

```python
def my_print (a, b) :
    print("왼쪽:", a)
    print("오른쪽:", b)

my_print(b=100, a=200)

# 왼쪽: b=100
# 오른쪽: a=200
# 틀렸다, 순서가 바뀌어도 지정한 a와 b를 읽는다!

# 왼쪽 200
# 오른쪽 100
```
