```python
data = [
    [ 2000,  3050,  2050,  1980],
    [ 7500,  2050,  2050,  1980],
    [15450, 15050, 15550, 14900]
]

for row in data:
    for col in row:
        print(col * 1.00014) #수수료 0.014% 이므로 0.00014
```

```python
for row in data:
    print(row[0] * 1.00014)
    print(row[1] * 1.00014)
    print(row[2] * 1.00014)
    print(row[3] * 1.00014)#수수료 0.014% 이므로 0.00014
    print("-----")
```

```python
for row in data:
    for col in row:
        print(col * 1.00014) #수수료 0.014% 이므로 0.00014
    print("-----") #따란
```

```python
result = []
for row in data:
    for col in row:
        result.append(col * 1.00014) #수수료 0.014% 이므로 0.00014
print(result)
```

```python
result = []

for row in data:
    sub = []
    for col in row:
        
        sub.append(col * 1.00014)
    result.append(sub)#수수료 0.014% 이므로 0.00014
print(result)
```

```python
ohlc = [["open", "high", "low", "close"],
        [100, 110, 70, 100],
        [200, 210, 180, 190],
        [300, 310, 300, 310]]
close = []
for row in ohlc:
    if type(row[3])==str: 
        None
    else:
        close.append(row[3])
print(close)
```

```python
for row in ohlc[1::]: #인덱스0 빼고
    print(row[3])
```

```python
for row in ohlc[1::]: #인덱스0 빼고
    if row[3] > 150:
        print(row[3])
```

```python
for row in ohlc[1::]:
    if row[0] <= row[3]:
        print(row[3])
```

```python
volatility = []
for row in ohlc[1::]: #인덱스0 빼고
    volatility.append(row[1] - row[2])
print(volatility)
```

```python
volatility = []
for row in ohlc[1::]: #인덱스0 빼고
    if row[0] < row[3]:
        volatility.append(row[1] - row[2])
    else: None
print(volatility)
```

```python
for row in ohlc[1::]:
    if row[3] > row[0]:
        print(row[1] - row[2])
```
